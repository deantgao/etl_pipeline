```python
data_files = []
def retrieve_files(directory):
    for folder in os.listdir(directory):
        new_directory = directory + "/" + folder
        if os.path.isfile(new_directory):
            data_files.append(new_directory)
        if os.path.isdir(new_directory):
            for item in os.listdir(new_directory):
                if not item.endswith('.Rhistory'):
                    if item.endswith('.sas7bdat') or item.endswith('.pdf') or item.endswith('.csv') or item.endswith('.TXT') or item.endswith('.txt') or item.endswith('.xlsx') or item.endswith('.xlsm') or item.endswith('.xls'):
                        filepath = new_directory + "/" + item.lower()
                        data_files.append(filepath)
                    else:
                        retrieve_files(new_directory+"/"+item)               
    return data_files
```

```python
def create_lists_of_cat(data_files):
    eligibility_files = []
    randomization_files = []
    subject_visit_files = []
    demog_files = []
    for each in data_files:
        if 'udp' not in each[1] and ('elig' in each[1] or 'Elig' in each[1] or 'ELIG' in each[1] or 'iex' in each[1] or '461-018/dta_sbj_participation' in each[1]):
            eligibility_files.append(each[1])
        if 'visit' in each[1] or 'Visit' in each[1] or 'xdov' in each[1] or '461-013/sv' in each[1]:
            subject_visit_files.append(each[1])
        if 'rand' in each[1] or 'Rand' in each[1] or 'myk461008_enrollment_20200604010001' in each[1] or '461-015/enrol.sas7bdat' in each[1] or '461-012/dta_sbj' in each[1] or '461-014/dta_sbj' in each[1]:
            randomization_files.append(each[1])
        if 'dm' in each[1] or 'DEMO' in each[1] or 'demo' in each[1]:
            demog_files.append(each[1])
    return eligibility_files, randomization_files, subject_visit_files, demog_files
```

```python
def make_visible_index(data_files):
    for i in range(len(data_files)):
        data_files[i] = (i,data_files[i])
    return data_files
```

```python
def run_all_for_lists_by_cat(directory):
    files = retrieve_files(directory)
    indexed_files = make_visible_index(files)
    elig, rand, visit, demog = create_lists_of_cat(files)
    return indexed_files, elig, rand, visit, demog
```
