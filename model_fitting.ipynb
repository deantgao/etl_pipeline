```python
def fill_ex_id_10(df):
    for i,row in df.iterrows():
        if 'criteria #10' in row.reason_for_sf.lower():
            df.loc[i,'ex_id'].pop()
            df.loc[i,'ex_id'].append(10)
    return df
```

```python
def clean_protocol_num_col(df):
    df['protocol_num'] = df.protocol_num.fillna('1')
    for i,row in df.iterrows():
        if 'ORIGINAL' in row.protocol_num or 'AMENDMENT' in row.protocol_num:
            df.loc[i,'protocol_num'] = '1'
    return df

def pull_all_in_ex_codes(df):
    in_criteria = {}
    ex_criteria = {}
    for i,row in df.iterrows():
        if row.study_id not in in_criteria:
            in_criteria[row.study_id] = {row.protocol_num:{}}
        in_criteria[row.study_id][row.protocol_num] = {}
        if row.study_id not in ex_criteria:
            ex_criteria[row.study_id] = {row.protocol_num:{}}
        ex_criteria[row.study_id][row.protocol_num] = {}
#         print(i,'in_crit',in_criteria)
        for code in row.in_id:
#             print('protocol: ',row.protocol_num)
#             print(i, 'in code: ',code)
            if code not in in_criteria[row.study_id][row.protocol_num]:
                in_criteria[row.study_id][row.protocol_num][code] = ''
            in_criteria[row.study_id][row.protocol_num][code] += ''
#             print('study_id: ',row.study_id)
#             print('protocol num: ',row.protocol_num)
        ex_criteria[row.study_id] = {row.protocol_num:{}}
        for code2 in row.ex_id:
            if code not in ex_criteria[row.study_id][row.protocol_num]:
                ex_criteria[row.study_id][row.protocol_num][code2] = ''
            ex_criteria[row.study_id][row.protocol_num][code2] += ''
    return in_criteria, ex_criteria

SCREEN_FAILS1 = clean_protocol_num_col(SCREEN_FAILS_in_ex)
in_criteria, ex_criteria = pull_all_in_ex_codes(SCREEN_FAILS_in_ex)
```

```python
# This function IDs sites that participated in at least 2 (or more) studies
def id_sites_mult_studies(sf_training_data):
    sites_mult_studies = []
    for each in sf_training_data.groupby(['site_id']):
        sites_studies = []
        for i,row in each[1].iterrows():
            if row.study_id not in sites_studies:
                sites_studies.append(row.study_id)
        if len(sites_studies) > 1:
            sites_mult_studies.append(each[1].site_name.values[0])
    return sites_mult_studies

# A sorted list of all the sites that participated in at least two studies
sites_mult_studies = id_sites_mult_studies(patient_data)
sorted(sites_mult_studies)
```

```python
def clean_age_80(df,in_criteria,ex_criteria):
    for i,row in df.iterrows():
        df.loc[i,'reason_for_sf'] = ''
        for k in range(len(row.in_id)):
            if k > 0:
                df.loc[i,'reason_for_sf'] += '; '
            df.loc[i,'reason_for_sf'] += in_criteria[row.study_id][row.protocol_num][row.in_id[k]]
        for z in range(len(row.ex_id)):
            if df.loc[i,'reason_for_sf']=='':
                df.loc[i,'reason_for_sf'] += '; '
            df.loc[i,'reason_for_sf'] += ex_criteria[row.study_id][row.protocol_num][row.ex_id[z]]
    df = df.drop(['reason_for_sf2'],axis=1)
    return df

clean_age_80 = clean_age_80(age_80_sf_records,in_criteria,ex_criteria)
```

```python
def clean_race_records(df,in_criteria,ex_criteria):
    df = df.fillna('')
    df = df.drop(['reason_for_sf2'],axis=1)
    for i,row in df.iterrows():
        if df.loc[i,'reason_for_sf']=='':
            for each in row.ex_id:
                df.loc[i,'reason_for_sf'] = ex_criteria[row.study_id][row.protocol_num][each]
            for each2 in row.in_id:
                df.loc[i,'reason_for_sf'] = in_criteria[row.study_id][row.protocol_num][each2]
    return df
        
clean_race_black = clean_race_records(black_sf_records,in_criteria,ex_criteria)
clean_race_white = clean_race_records(white_sf_records,in_criteria,ex_criteria)
```

```python
def clean_phase_records(df):
    to_remove = ['inclusion','criteria','critiera','exclusion','-']
    digits = '0123456789'
    df = df.fillna('')
    for i,row in df.iterrows():
        if row.study_id == 'myk_224_001':
            words = row.reason_for_sf.split(' ')
            reason_for_sf = [word for word in words if not word in to_remove]
            reason_for_sf = [w for w in reason_for_sf if "#" not in w]
            reason_for_sf = [no_colon for no_colon in reason_for_sf if ':' not in no_colon]
            reason_for_sf = ' '.join(reason_for_sf)
            reason_for_sf = reason_for_sf.lstrip(digits)
            reason_for_sf = reason_for_sf.lstrip(';')
            reason_for_sf = reason_for_sf.rstrip('.')
            reason_for_sf = reason_for_sf.strip()
            if 'abnormal ecg' in reason_for_sf and '110' in reason_for_sf:
                reason_for_sf = 'abnormal ecg; qrs >110'
            if 'history of drug' in reason_for_sf.lower():
                reason_for_sf = 'history of drug use'
            df.loc[i,'reason_for_sf'] = reason_for_sf
        if row.reason_for_sf == '':
            for k in range(len(row.in_id)):
                if k>0:
                    df.loc[i,'reason_for_sf'] += '; '
                df.loc[i,'reason_for_sf'] += in_criteria[row.study_id][row.protocol_num][row.in_id[k]]
            for z in range(len(row.ex_id)):
                if z>0:
                    df.loc[i,'reason_for_sf'] += '; '
                df.loc[i,'reason_for_sf'] += ex_criteria[row.study_id][row.protocol_num][row.ex_id[z]]
    df = df.drop(['reason_for_sf2'],axis=1)
    return df
    
clean_phase1 = clean_phase_records(phase1_sf_records)
```
